```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Criando o DataFrame com os dados fornecidos
dados = {
    "Data": ["15/01/2025"] * 5,
    "Cidade": ["São Paulo", "Rio de Janeiro", "Curitiba", "Porto Alegre", "Salvador"],
    "Temperatura Máxima (°C)": [30.5, 35.0, 24.0, 28.0, 31.0],
    "Temperatura Mínima (°C)": [22.0, 25.0, 18.0, 20.0, 24.5],
    "Precipitação (mm)": [12.0, np.nan, 8.0, 15.0, np.nan],
    "Umidade Relativa (%)": [78, 70, np.nan, 82, 80]
}

df = pd.DataFrame(dados)

# Substituir valores ausentes na coluna Precipitação pela média da coluna
media_precipitacao = df["Precipitação (mm)"].mean()
df["Precipitação (mm)"] = df["Precipitação (mm)"].fillna(media_precipitacao)

# Substituir valores ausentes na coluna Umidade Relativa pela mediana da coluna
mediana_umidade = df["Umidade Relativa (%)"].median()
df["Umidade Relativa (%)"] = df["Umidade Relativa (%)"].fillna(mediana_umidade)

# Criar a coluna Amplitude Térmica
df["Amplitude Térmica"] = df["Temperatura Máxima (°C)"] - df["Temperatura Mínima (°C)"]

# Criar um novo DataFrame com cidades cuja Temperatura Máxima seja maior que 30°C
df_filtrado = df[df["Temperatura Máxima (°C)"] > 30]

# Reordenar as colunas
df = df[["Data", "Cidade", "Temperatura Máxima (°C)", "Temperatura Mínima (°C)", "Amplitude Térmica", "Precipitação (mm)", "Umidade Relativa (%)"]]
df_filtrado = df_filtrado[["Data", "Cidade", "Temperatura Máxima (°C)", "Temperatura Mínima (°C)", "Amplitude Térmica", "Precipitação (mm)", "Umidade Relativa (%)"]]

# Exibir os DataFrames
print("DataFrame Original:")
print(df)
print("\nDataFrame Filtrado:")
print(df_filtrado)
```

```python
# Gráfico de linha - Evolução da Temperatura ao longo do dia
horas = list(range(25))  # De 0 a 24 horas
temperaturas = [15 + (h * 1.25) if h <= 12 else 30 - (h - 12) for h in horas]  # Simulação de variação

plt.figure(figsize=(10, 5))
plt.plot(horas, temperaturas, marker='o', linestyle='-', color='b', label='Temperatura (°C)')
plt.xlabel("Hora do dia")
plt.ylabel("Temperatura (°C)")
plt.title("Evolução da Temperatura ao longo do dia")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Simulação de dados de vendas para análise com Seaborn
dados_vendas = {
    "Dia": ["Seg", "Ter", "Qua", "Qui", "Sex", "Sáb", "Dom"],
    "Vendas Totais": [200, 250, 180, 220, 300, 400, 380],
    "Clientes": [50, 65, 45, 60, 80, 120, 110],
    "Lucro": [500, 700, 450, 600, 900, 1500, 1400]
}
df_vendas = pd.DataFrame(dados_vendas)

# Gráfico de barras - Vendas Totais por dia
plt.figure(figsize=(8, 5))
sns.barplot(x="Dia", y="Vendas Totais", data=df_vendas, hue="Dia", palette="viridis", legend=False)
plt.title("Vendas Totais por Dia")
plt.xlabel("Dia da Semana")
plt.ylabel("Vendas Totais")
plt.show()

# Gráfico de dispersão - Clientes vs Vendas Totais
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Clientes", y="Vendas Totais", data=df_vendas, color="red")
plt.title("Relação entre Clientes e Vendas Totais")
plt.xlabel("Número de Clientes")
plt.ylabel("Vendas Totais")
plt.show()

# Heatmap de correlação (removendo a coluna categórica "Dia")
df_corr = df_vendas.drop(columns=["Dia"]).corr()
plt.figure(figsize=(6, 5))
sns.heatmap(df_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Calor - Correlação entre Variáveis")
plt.show()
```
