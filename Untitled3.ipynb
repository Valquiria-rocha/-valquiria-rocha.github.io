```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Gerando dados fictícios
np.random.seed(42)
engine_power = np.random.randint(50, 300, 100)  # Potência do motor (cv)
vehicle_weight = np.random.randint(800, 2500, 100)  # Peso do veículo (kg)
economy = (engine_power / vehicle_weight < 0.08).astype(int)  # Classificação

X = np.column_stack((engine_power, vehicle_weight))
y = economy

# Criando e treinando o modelo
tree = DecisionTreeClassifier(max_depth=3, random_state=42)
tree.fit(X, y)

# Visualizando a árvore de decisão
plt.figure(figsize=(10, 6))
plot_tree(tree, feature_names=['Potência do Motor', 'Peso do Veículo'], class_names=['Não Econômico', 'Econômico'], filled=True)
plt.title('Árvore de Decisão para Classificação de Veículos')
plt.show()
```
